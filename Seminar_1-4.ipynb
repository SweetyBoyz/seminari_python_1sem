```python
my_str = "one dog, one cat, one rabbit"

#Заменяем все вхождения "one" в строке
a = my_str.replace("one", "two")
print(a) # Выведет two dog, two cat, two rabbit

#Заменяем первое вхождение "one" в строке
b = my_str.replace("one", "two", 1)
print(b) # Выведет two dog, one cat, one rabbit

#Заменяем первые два вхождения "one" в строке
c = my_str.replace("one", "two", 2)
print(c) # Выведет two dog, two cat, one rabbit
```

### Функция работы со строками split

string.split(separator,max)

- разделитель (необязательный параметр) – строка разбивается на части с помощью указанного символа. Если разделитель не задан, то любой симво, строка (пробел, новая строка и т.д.) считается 
разделителем;

- maxsplit (необязательный параметр) определяет максимальное количество частей.


```python
grocery = 'Milk, Chicken, Bread, Butter'
# maxsplit: 2
print(grocery.split(', ', 2))
# maxsplit: 1
print(grocery.split(', ', 1))
# maxsplit: 5
print(grocery.split(', ', 5))
# maxsplit: 0
print(grocery.split(', ', 0))
```

# Семинар 3. Словари

### Словари в Python - неупорядоченные коллекции произвольных объектов с доступом по ключу. Их иногда ещё называют ассоциативными массивами или хеш-таблицами.

```python
'''Создание пустого словаря:'''
d = {}

'''Cловарь, где ключи являются целыми числами:'''
d = {1:'dict', 2:'dictionary'}
print(d)

'''Создание словаря с ключами разных типов'''
dict_sample = {'fruit': 'mango', 1: [4, 6, 8]}
print(dict_sample)

'''========================Можно также создать словарь, явно вызвав метод dict()'''
dict_sample = dict({1:'mango', 2:'pawpaw'})
print(dict_sample)


'''fromkeys(seq[, value]) -> dict Новый словарь, созданный из последовательности.
Возвращает новый словарь, сформированный из указанной последовательности,
с указанным значением элементов.'''
d = dict.fromkeys(['a', 'b', 1], 100)
print(d)
print(d.get('a'))

'''================генератор словарей============================='''
d = {a: a ** 2 for a in range(7)}
print(d)
```

### Методы словарей
clear() – очищает заданный словарь, приводя его к пустому.

get() – отдаёт значение словаря по указанному ключу. Если ключ не существует, а в качестве
дополнительного аргумента передано значение по умолчанию, то метод вернет его. Если же значение по умолчанию опущено, метод вернет None.

items() – возвращает словарные пары ключ:значение, как соответствующие им кортежи.

keys() – возвращает ключи словаря, организованные в виде списка.

values() – подобным образом, возвращает список значений словаря.

pop() – удалит запись словаря по ключу и вернет её значение.

popitem() – удаляет последний элемент в словаре.

update() – реализует своеобразную операцию конкатенации для словарей. Он объединяет ключи и значения одного словаря с ключами и значениями другого. При этом если какие-то ключи совпадут, то результирующим значением станет значение словаря, указанного в качестве аргумента метода update.

copy() – создает полную копию исходного словаря.

```python
'''dict.clear() - очищает словарь'''
d = {a: a  for a in range(7)}
print(d)
d.clear()
print(d)

'''dict.copy() - возвращает копию словаря'''
dict_sample = {
  "Company": "Toyota", 
  "model": "Premio", 
  "year": 2012 
} 
x = dict_sample.copy() 
print(x)

'''dict.get(key[, default]) - обращение к элементу словаря по ключу'''
story_count = {'сто': 100, 'девяносто': 90, 'двенадцать': 12, 'пять': 5}
y = story_count.get('двенадцать')
print(y)

'''dict.items() - возвращает пары (ключ, значение), итерация'''
dict_sample = { "Company": "Toyota", "model": "Premio",  "year": 2012 } 
for k, v in dict_sample.items(): 
    print(k, v)

'''dict.keys() возвращает ключи словаря, организованные в виде списка'''
dict_sample = {1:'mango', 2:'pawpaw'}    
x = dict_sample.keys() 
print(x)    
    
'''values() – возвращает список значений словаря'''
dict_sample = {1:'mango', 2:'pawpaw'}    
x = dict_sample.values() 
print(x)  

''' pop() – удалит запись словаря по ключу '''
dict_sample = {
  "Company": "Toyota", 
  "model": "Premio", 
  "year": 2012 
} 
dict_sample.pop("year") 
print(dict_sample)

'''popitem() удаляет последний элемент в словаре.'''
another_dict = {'t': 16, 'g': 53, 'y': 112, 'h': 23, 'q':1}
print(another_dict.popitem())
print(another_dict)

'''update() объединение словарей'''
showcase_1 = {'Apple': 2.7, 'Grape': 3.5, 'Banana': 4.4}
showcase_2 = {'Orange': 1.9, 'Coconut': 10}
showcase_1.update(showcase_2)
print(showcase_1)
```

# Семинар 4. Множества
Множество - это неупорядоченная коллекция уникальных элементов, с которой можно сравнивать другие элементы, чтобы определить, принадлежат ли они этому множеству. Множество может содержать только элементы неизменяемых типов, например числа, строки, кортежи. Объявить множество можно с помощью конструктора set().

```python
'''В Python 3 можно также создать множество,
указав элементы внутри фигурных скобок. Обратите внимание на то,
что при указании пустых фигурных скобок будет создан словарь, а не множество.
Чтобы создать пустое множество, следует использовать функцию set().'''
s1 = set()
s2 = {1,2,3,4}
s3 = {}
print('s1 = ', s1, ' type s2 = ', type(s2), ' type s3 = ', type(s3))
```

```python
'''в set() можно передать какой-либо объект, по которому можно проитерироваться (Iterable):'''
# преобразуем список в множество:
s2 = set ([1, 2, 3, 4, 5])
print("s2 = ", s2)
'''Элементы множества должны быть уникальными, множество не может содержать одинаковых элементов.
Добавление элементов, которые уже есть в множестве, не изменяет это множество.'''
s3 = set ([1, 2, 3, 4, 5, 1, 2])
print("s3 = ", s3)
```

```python
'''это использование set comprehension. Это специальная синтаксическая конструкция языка,
которую иногда называют абстракцией множества по аналогии с list comprehension (Списковое включение).
это генератор множеств'''
numbers = [1, 2, 2, 2, 3, 3, 4, 4, 5, 6]

even_numbers = {number for number in numbers }
print('even_numbers =',even_numbers)
set_numbers = {number for number in range(1,10,2) }
print('set_numbers = ', set_numbers)
```

```python
# обход элементов множества в цикле for:
s11 = set(['banana', 'apple', 'orange'])
for v in s11:
    print(v, end=' ')
print('\n','length s11 = ', len(s11))
```

```python
s3 = {1, 2, 3, 4}
s4 = {2, 4, 5, 6}
# объединение множеств:
s3.union(s4)
```

```python
# добавляют элементы множества s4 во множество s5:
s41 = {2, 4, 5, 6}
s5 = {7, 8 , 9}
s5.update(s41) '''эквивалентно: s5 |= s4''' 
s5
```

```python
s51 = {2, 4, 5, 6}
s52 = {7, 8 , 9}
s52 |=s51 
s52
```

```python
'''а &= b и а.intersection(b)
- во множестве а останутся элементы, которые существуют и во множестве а, и во множестве b'''
s6 = {2, 4, 5, 6}
s7 = {7, 8 , 9, 6}
s6 & s7
```

```python
s61 = {2, 4, 5, 6}
s71 = {7, 8 , 9, 6}
print(s61.intersection(s71))
```

```python
# разница множеств:
s4 - s3 # s4.difference(s3)
```

```python
# удаляем элементы из множества а, которые присутствуют во множестве b:
s6 = {2, 4, 5, 6} 
s6 -= s3 # s6.difference_update(s3)
s6
```

```python
'''^ и symmetric_difference()
- возвращают элементы обоих множеств, присутствующие только в одном из множеств-аргументов.'''
{3, 4, 5, 6} ^ {5, 6, 7, 8}
```

```python
'''in - проверка наличия элемента во множестве:'''
3 in {1,2,3,4,5}
```

```python
'''in - проверка наличия элемента во множестве:'''
set([1, 2]) == set([1, 2, 3])
```

а <= b и a.issubset(b) - проверяют, входят ли все элементы множества а во множество b

а < b - проверяет, входят ли все элементы множества а во множество b. Причем множество а не должно быть равно множеству b

а >= b и а.issuperset(b) - проверяют, входят ли все элементы множества b во множество а

а > b - проверяет, входят ли все элементы множества b во множество а. Причем множество а не должно быть равно множеству b

а.isdisjoint(b) - возвращает True, если результатом пересечения множеств а и b является пустое множество (это означает, что множества не имеют одинаковых элементов)

сору() - создает копию множества. Обратите внимание на то, что оператор = присваивает лишь ссылку на тот же объект, а не копирует его.

add(<Элемент>) - добавляет <Элемент> во множество

remove(<Элемент>) - удаляет <Элемент> из множества. Если элемент не найден, то возбуждается исключение KeyError

disсаrd(<Элемент>)- удаляет <Элемент> из множества, если он присутствует

clear() - удаляет все элементы из множества

```python
s8 = {1, 2, 3}
s8.add(4)
s8
```

```python
s9 = {1, 2, 3}
s9.remove(2)
s9
```

```python
'''рор() - удаляет произвольный элемент 
из множества и возвращает его. Если элементов нет, то возбуждается исключение KeyError'''
s11 = {1, 2, 3}
s11.pop()
```
